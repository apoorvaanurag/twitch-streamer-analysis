## peak_avg_diff_pc

```python
twitch_df['peak_avg_diff_pc'] = twitch_df['peak_viewer_pc'] - twitch_df['avg_viewer_pc']
```

```python
twitch_df.head()
```

### Describing the calculated metrics
* time_score : This ratio of stream time to watch time is significant to gauge the popularity of a particular channel. A higher ratio implies simultaneous viewers watched content for a longer time than it was streamed.

* avg_viewer_pc : Since average viewers give us an idea of central tendency, this statistic helps us compare the total followers with the viewers across the year. 

* peak_viewer_pc :  A ratio of peak viewers at any moment to the total followers of the channel

* peak_avg_diff_pc : Difference between avg_viewer_pc and peak_viewer_pc as a measure of consistency in terms of viewers


```python
streamer_metrics = twitch_df[['Channel','time_score', 'avg_viewer_pc', 'peak_viewer_pc', 'peak_avg_diff_pc', 'Partnered', 'Mature','Language']].copy()
```
